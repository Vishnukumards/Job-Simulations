```python
InvoiceNumber = []
for i in data["InvoiceNo"]:
    if isinstance(i, int):
        InvoiceNumber.append(i)
    elif isinstance(i, str):
        InvoiceNumber.append(int(''.join(filter(str.isdigit, i))))
```

```python
data["InvoiceNo"]= InvoiceNumber
```

```python
new_quantity_list = []
average_quantity = int(data["Quantity"].mean())
for i in data["Quantity"]:
    if i < 1:
        new_quantity_list.append(average_quantity)
    else:
        new_quantity_list.append(i)

data["Quantity"] = new_quantity_list
```

```python
# Create a check that the quantity should not be below 1 unit

if (data["Quantity"] < 1).any() :
    print("yes")
else :
    print("no")
```

```python
# Create a check that the Unit price should not be below $0

if (data["UnitPrice"] < 1).any() :
    print("yes")
else :
    print("no")
```

```python
data["UnitPrice"].mean()
```

```python
new_UnitPrice_list = []
average_UnitPrice = int(data["UnitPrice"].mean())
for i in data["Quantity"]:
    if i < 1:
        new_UnitPrice_list.append(average_UnitPrice)
    else:
        new_UnitPrice_list.append(i)

data["UnitPrice"] = new_UnitPrice_list
```

```python
if (data["UnitPrice"] < 1).any() :
    print("yes")
else :
    print("no")
```
